## 1. Introduction ##

**Production planning** is essential for the management of manufacturing to produce the right number of products to satisfy customer demand over a specific time horizon and maximize profit. <br><br>
**Brief History:** <br>
The origin of production planning is important to its development today. According to Rondeau(2001), when the first manufacturing planning and control system came out to the public, it consisted of groups of plant foremen and each one of them was responsible for scheduling production, ordering materials and shipping products to their assigned plant area. Even though the employees have different skills,the relatively simple design and operation of these plants allowed even the most poorly trained foreman to operate them successfully. As these early industrial era firms evolved, more advanced production and inventory control gradually replaced the simple and often inefficient plant foreman system. <br><br>
Later in the 1970s, the primary competitive thrust shifted to marketing, which resulted in the adoption of target-market strategies which emphasized on greater production integration and planning. Materials Requirement Planning (MRP) systems fit that requirement nicely. In the 1980s the primary competitive thrust shifted to quality. This time the manufacturing strategy emphasized on having more advenced process control, greater world class manufacturing, and reducing overhead costs (p.1). This topic is important because with an efficient product planning model, factories can produce their products more efficiently, that they could not only increase production but also reduce overhead cost such as direct labor and direct materials. <br><br>
For stochastic production, it is also essential to know how it was created. According to Yin, Yin and Liu (2004), when outcomes are not determined, discrete‚Äêtime Markov Decision Processes (MDP) introduced in the 1950s, which use stochastic dynamic programming, are models for sequential decision making. In the past decades, MDPs have become an important research area in both theory and applications. MDP models have been used to solve many real‚Äêworld problems in areas such as ecology, economics, communication, electrical engineering, and computer science (p.1). Knowing stochastic production is important because not only we need to know another way of production planning but also this model has been applied to various fields. <br><br>
**Goal:**<br>
Our final project aims to match production and sourcing decisions to meet market demand subject to production capacity, workforce availability, and overtime restrictions. 
The objective of the problem is to maximize the profit or minimize the total cost. The best situation is knowing future demand, a deterministic problem. In this situation, we will use deterministic production planning model to solve the problem. However, in the real world, manufacturing is a complex dynamic system with many uncertain and random demands, which makes planning difficult to manage. We use stochastic production planning model to achieve the goal. 
<br><br>
**The two types of mathematical models are:**<br>
#### A: Deterministic Production Planning Model<br>
In this model, the demand, $d$, for a period of time, $t$, is known before manufacturing. The decision variables are the production quantity of the product, the inventory level at the beginning of time $t$, and the indicator (binary variable) of prodcing status in time $t$. In each period of time, we are given the cost of workforce to produce the product, the cost to produce each unit of product, the storage cost, the manufacturing capacity, and the demand. We model and solve the production planning problem by minimizing the total cost. <br>
#### B: Stochastic Production Planning Model<br>
Here, the demand of a product is uncertain. However, manufacturing can be carried out by using best guess demands. We can have a set of probabilities associated with the possible demand to find a solution that is feasible for all or almost all the possible data, and then optimize the expected performance of the model.

## 2. Mathematical models ##

### Model A: Deterministic Production Planning (LP & MIP)
 
   1. **Decision variables**
      * $x(t)$ is the production produced in a period time $t$
      * $ùë¶(t)$ is the inventory level at the end of each period of time $t$
      * $ùëß(t) \in \{0, 1\}$
         * 1 if production occurs in time $t$
         * 0 otherwise<br><br>
    
   2. **Parameters**
      * $f(t)$ is the workforce cost of producing in time $t$
      * $c(t)$ is the cost of a unit of production in time $t$
      * $h(t)$ is the cost of the storage in time $t$
      * $C(t)$ is the capacity of manufacturing in time $t$
      * $d(t)$ is the demand of the product in time $t$<br><br>
      
   3. **Constraints**
      * Nonnegativity and integer constraints
          * $ x(t), \ y(t)  \ \geq 0 \text{ and } z(t) \ \in \{0, 1\}, \ \forall t \in \{1, 2,..., n\}$ <br><br>
    
      * Capacity constraint
          * $ x(t) \ \leq  \ Cz(t), \ \forall t \ \in \ {1, 2, ..., n}$<br><br>

      * Conservation constraint
          * $ y(t) \ + \ x(t) \ = \ d(t) \ + \ y(t + 1), \ \forall t \in \{1,2,...,n\}$<br><br>
          
   4. **Objective**
    <br>
    $$
    \begin{align}
    \text{Minimize} \qquad
    & \displaystyle\sum_{t \ = \ 1}^n c(t)x(t) \ + \ \displaystyle\sum_{t \ = \ 1}^n f(t)z(t) \ 
    + \ \displaystyle\sum_{t \ = \ 1}^n h(t)y(t) \\
    \text{Subject to} \qquad
    & x(t), y(t),  \ \geq  \ 0, \ z(t) \in \{0, \ 1\}, \\
    & y(t) \ + \ x(t) \ = \ d(t) \ + \ y(t + 1), \ \forall t \in \{1,2,...,n\}
    \end{align}
    $$
 <br>
 
###  Model B:  Stochastic Production Planning Model (MIP)<br>
 
   1. **Decision variables**
    * $r_1,r_2,‚Ä¶,r_m$ are the amounts of raw materials requried to produce $n$ different products, where $i \ = \ 1, 2, ..., m$ and $r_i \geq 0$.
    * $q_1, q_2,‚Ä¶, q_n$ are the quantities of $n$ different products, where $q_j \geq 0$.<br><br>
     
   2. **Parameters**
    * $ A_{ij}$, manufacturing one unit of product $j$ requires $A_{ij}$ units of raw material $i$, where $r \geq Aq$ and $A$ is nonnegative.
    * $ c \in R_{+}^m$, is the cost for raw material $r$. The total cost is $c^Tr$.
    * $ p \in R_{+}^n$, is the vector of product prices. The total profit is $p^Ts \ - \ c^Tr$.
    * $ \pi_1,\pi_2,...,\pi_k$, are the probabilities of a set of $K$ possible demand vectors $d^{(1)},...,d^{(k)}$, where $1^T\pi = 1, \ \pi \geq 0 $.
    * $s_j \ = \ min\{q_j, \ d_j\}$, is the number of units of product $j$ sold, where $ s_j \geq 0$.
        * If $q_j \gt d_j, \ q_j - d_j $ is the amount of product $j$ produced but not sold. 
        * If $q_j \lt d_j, \ q_j - d_j $ is the amount of unmet demand.
    * $C$ is the manufacturing capacity.<br><br>
    
   3. **Constraints**
    * $ r \geq Aq$, because manufacturing one unit of product $j$ requires at least $A_{ij}$ units of
      raw material $i$.
    * $q \geq 0, \ r\geq 0$. <br><br>
    
   4. **Objective**
    * **Case I**: Choose $r$ and $q$ before the demand $d$ is known<br>
         We incorporate the probabilities of demand into the model and maximize the expected profit.
         $$
         \begin{align}
         \text{Maximize} \qquad
         & -c^Tr \ + \ \sum_{k \ = \ 1}^K \pi_kp^T min\{q,\ d^k\} \\
         \text{Subject to} \qquad
         & r,q \geq 0, \ r \geq Aq, \ k \ =1,...,K
         \end{align}
         $$
         <br><br>
    * **Case II**: Choose $r$ from case I, and then $q$ after $d$ is known<br>
    In this case we have variables $r \ \in \ R_{+}^m \ $ and $ \ q^k \in R_{+}^n, \ k \ = \ 1, ...,K,$ where $q^k$ is the product we produce if $d^k$ turns out to be the actual product demand. <br>
    Then, the objective is to maximize the expected profit.<br>
    $$
    \begin{align}
    \text{Maximize} \qquad
    & -c^Tr  \ + \ \sum_{k \ = \ 1}^K \pi_kp^Tq^k \\
    \text{Subject to} \qquad
    & r, d^k, q^k \geq 0, \ r \geq Aq^k, \ k \ = \ 1,...K
    \end{align}
    $$
  

## 3. Solution ##
#### This solution section implements the two models of the optimization problems.
 - The first model solves the production planning problem with two senarios of known demands.
 - The second model solves the production planning problem with unknown demands from two approaches.

### A. Deterministic Production Planning
We are given vectors of workforce cost, unit production cost, storage cost, demands, and maufacturing capacity in time $t = 1, 2, ..., 24$ (month). The product quantity produced in time $t$, the inventory level at the beginning of the time $t$, and production indicator $z$ (1 is on 0 otherwise) are variables to find the optimal solution by minimizing the total cost. 

```julia
# Data
raw = readcsv("dpp.csv")

# f(t) -- the workforce cost of producing in time t
f = raw[:, 1][:]

# c(t) -- the cost of a unit of production in time t
c = raw[:, 2][:]

# h(t) -- the cost of the storage in time t
h = raw[:, 3][:]

# C -- the capacity of manufacturing
C = raw[:, 4][:]

# d(t) -- the demand of the product in time t
d1 = raw[:, 5][:]
d2 = raw[:, 6][:]

# Production in a year, i.e. twelve months
t = length(d1)
```

```julia
# Deterministic Production Planning Model
using JuMP, Cbc
m = Model(solver = CbcSolver())

# Using First Demand set

@variable(m, x[1:t] >= 0)   # production produced in time t
@variable(m, y[1:t+1] >= 0) # the inventory level at the end of time t
@variable(m, z[1:t], Bin)   # 1 if production occurs in time t, 0 otherwise
@constraint(m, x .<= C.*z)  # production constraints
@constraint(m, y[t+1] == 0) # clear out inventory at the end of production planning

# conservation constraint
@constraint(m, flow[i in 1:t], y[i] + x[i] == d1[i] + y[i+1])
@expression(m, material_cost, sum(c[i].*x[i] for i = 1:t))
@expression(m, storage_cost, sum(h[i]*y[i] for i = 1:t))
@expression(m, labor_cost, sum(f[i]*z[i] for i = 1:t))
@objective(m, Min, material_cost + storage_cost + labor_cost)

status1 = solve(m)

xx1 = getvalue(x)
yy1 = getvalue(y)
zz1 = getvalue(z)
ob1 = getobjectivevalue(m)

# Using second demand set
m = Model(solver = CbcSolver())

@variable(m, x[1:t] >= 0)   # production produced in time t
@variable(m, y[1:t+1] >= 0) # the inventory level at the end of time t
@variable(m, z[1:t], Bin)   # 1 if production occurs in time t, 0 otherwise
@constraint(m, x .<= C.*z)  # production constraints
@constraint(m, y[t+1] == 0) # clear out inventory at the end of production planning

# conservation constraint
@constraint(m, flow[i in 1:t], y[i] + x[i] == d2[i] + y[i+1])
@expression(m, material_cost, sum(c[i].*x[i] for i = 1:t))
@expression(m, storage_cost, sum(h[i]*y[i] for i = 1:t))
@expression(m, labor_cost, sum(f[i]*z[i] for i = 1:t))
@objective(m, Min, material_cost + storage_cost + labor_cost)

status2 = solve(m)

xx2 = getvalue(x)
yy2 = getvalue(y)
zz2 = getvalue(z)
ob2 = getobjectivevalue(m)
```

### B.  Stochastic Production Planning 

### First Approach
In this approach, we use the possible demands with associated probabilities to set the upper bound of the production.

#### Case I: Choose $r$ and $q$ ahead, before the demand $d$ is known

In this case the demands are unknown. We solve the problem by setting the maximum producing quantities of different products to the possible demands with associated probabilities to prevent over producing. After solving the number of raw materials to purchase and the quantities of products to produce, we then compute the expected profit by using the possible sold quantities of products from different buyers. 
<br>

```julia
# Data
raw2 = readcsv("spp.csv")
(a, b) = size(raw2)
unit_material = 5:9
demand = 10:b

# M -- number of raw materials
M = 10

# N -- number of products
N = 5

# œÄ -- the probability of a set of the possible demand vectors
œÄ = raw2[:,1][:]

# K -- number of possibility vectors 
K = length(œÄ)

# c -- the cost for raw material r.
c = raw2[:,2][:]
c = c[1:M]

# p -- the vector of product prices
p = raw2[:,3][:]
p = p[1:N]

# C -- the manufacturing capacity
Cap = raw2[:,4][:]
Cap = Cap[1:N]

# A -- units of raw material i needed for one unit of product j
A = raw2[:, unit_material]
A = A[1:10,1:5]

# possible demand vectors associated with the probabilities
D = raw2[:, demand]
D = D'
```

```julia
# Undeterministic Production Planning Model Case I
using JuMP, Cbc, Gurobi
m = Model(solver = CbcSolver())

# stage one
# produce products based on possible demand
@variable(m, r[1:M] >= 0) # amount of each raw material to buy
@variable(m, q[1:N] >= 0) # amount of each product to manufacture
@constraint(m, r .>= A*q) # material constraint
@constraint(m, q .<= D*œÄ)
@constraint(m, q .<= Cap)
@expression(m, cost, c.*r)
@expression(m, revenue, sum(p[i]*q[i] for i = 1:N))
@objective(m, Max, -sum(cost) + revenue)
status1 = solve(m)

q1 = getvalue(q)
r1 = getvalue(r)
o1 = getobjectivevalue(m)

# statge two 
# we will use the r and q from stage one to compute expected profit
# Computes Expected Profit
R = ones(1, K)
S = zeros(N, M)
Q = q1*R
# sj = min{qj, dj} -- the number of units of product j sold
    # If  qj>dj, qj‚àídj -- the amount of product j produced but not sold.
    # If  qj<dj, qj‚àídj -- the amount of unmet demand.
S = min.(Q, D)
exp_pr = -sum(c.*r1) + sum(p'*q1)
profit = -sum(c.*r1) + sum(p'*S*œÄ)
```

#### Case II: Choose $r$ ahead of time, and $q$ after $d$ is known

In this case, we use the possible sold quantites computed from case I as production goal to find the proper amount of raw materials to purchase to avoid excess raw materials or produced products.

```julia
# Undeterministic Production Planning Model Case II
using JuMP, Cbc, Clp, Gurobi
m = Model(solver = CbcSolver())

@variable(m, r[1:M] >= 0)
@variable(m, q[1:N] >= 0) # product to manufacture for each demand
@constraint(m, r .>= A*q)
@constraint(m, q .== S*œÄ)
@expression(m, revenue, sum(p[i]*q[i] for i = 1:N))
@objective(m, Max, revenue - c'*r)

status2 = solve(m)

q2 = getvalue(q)
r2 = getvalue(r)
o2 = getobjectivevalue(m)
```

### Second Approach

In this approach, we generated a vector of five elements having values between one and two inclusive to set the upper bound and lower bound of production. 

#### Case I: Choose $r$ and $q$ ahead, before  the  demand $d$ is known

```julia
# generates a random vector with five elements 
temp1 = rand(1:2,1,N)

# Undeterministic Production Planning Model Case I

using JuMP, Cbc
m = Model(solver = CbcSolver())

@variable(m, r[1:M] >= 0) # amount of each raw material to buy
@variable(m, q[1:N] >= 0) # amount of each product to manufacture
@constraint(m, r .>= A*q) # material constraint
@constraint(m, q .<= Cap)

# randomly generates the quantity of each q
for i in 1:N
    if temp1[i] == 1
      #upper bound
      @constraint(m, sum(q[i]) >= sum(D[i,:] .* œÄ))
    else
      #upper bound
      @constraint(m, sum(q[i]) <= sum(D[i,:] .* œÄ))
      #lower bound
      @constraint(m, sum(q[i]) >= sum(D[i,:] .* œÄ *0.5)) 
    end
end

@expression(m, cost, c.*r)
@objective(m, Max, -sum(cost) + sum(p.*q))
solve(m)

qq1 = getvalue(q)
rr1 = getvalue(r)
oo1 = getobjectivevalue(m)
SS = min.(q1*ones(1,K), D)
```

#### Case II: Choose $r$ ahead of time, and $q$ after $d$ is known

```julia
using JuMP, Cbc, Gurobi
m = Model(solver = CbcSolver())
@variable(m, r[1:M] >= 0)
@variable(m, q[1:N] >= 0) #quantity of n products to be produced
@constraint(m, r .>= A*q2)    #material constraint
@constraint(m, q .<= Cap)   # capacity constraint
@constraint(m, q .== SS*œÄ) #use S_graph here to keep S from changing in order to have better discussion
@expression(m, cost, c.*r)   
@objective(m, Max, -sum(cost) + sum(p.*q))
solve(m)

qq2 = getvalue(q)
rr2 = getvalue(r)
oo2 = getobjectivevalue(m)
```

## 5. Conclusion ##

In both models, the objectives are piecewise-linear and the constraints are linear inequalities, so our production planning problems are concex optimization problems. The models are solved with randomly generated data.

The deterministic model is easy to solve. The constraints are all explicit. Knowing the future demand, we are able to build up the conservative constraint and to generate the optimize outcome.

In contrast, the stochastic model is more complex. We have tried multiple ways to build a reasonable model, many of them didn't work. At the begining, we tried to solve the problem with one stage by applying the variable q on the operation, $min(q, d^k)$, which is not supported by solvers. We then tried many other different ways and finally came up with two approaches that work properly. In both approaches, the results of the second case come to agreement. However, for the case one, the first approach is more reasonable. It might slightly under estimate production, yet, it is better compared to manufacuring products with over-estimated demands. The difference between expected profit and actual profit in second approach is about 38 times more than the difference in first appraoch. 

Therefore, we can apply this stochastic model in the real world by using the first case as the first estimation stage and the second case as the second estimation stage to generate overall results to help make a better decision of production planning with uncertainties.

It is unrealistic to expect the stochastic model to generate the best production plan, however, it is important to use stochastic model to avoid the worst production plan and find a good production plan.
